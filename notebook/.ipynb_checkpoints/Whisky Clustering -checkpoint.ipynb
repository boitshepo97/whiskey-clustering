# Clustering whisky distilleries according to tasting Profiles

Whiskey is a dark distilled spirit that is made from a variety of grains, including barley, corn, rye, and wheat. It is distilled throughout the world, most popularly in Ireland, Scotland, the United States, Canada, and Japan. There are various styles of whiskey and some countries have regulations that stipulate how it is produced. Whether it's Irish whiskey, scotch, bourbon, or Canadian whisky, it's the most popular liquor in the world and it's used in numerous cocktail and shot recipes. Surprisingly, drinkers in India consume the most whiskey, though the country's own whiskey is rarely exported.

#### What is K-Means?
K-means clustering is very sensitive to scale due to its reliance on Euclidean distance so be sure to normalize data if there are likely to be scaling problems.

## Flavours

```python
flavours.columns
```

## Whiskey Matrix

```python
sns.pairplot(flavours)
```

## Checking for correlation: Using Body as a hue.

Using a correlation heatmap, I will investigate how the flavours relate to the attribute "Body" this will further help me to determine what I need to do with the data.


```python
correlation = flavours.corr()
```

```python
# correlation heatmap
plt.subplots(figsize = (12,10))
sns.heatmap(correlation, vmin=-1, vmax=1, annot=True, linewidth=0.1)
```

## Grouping using correlation.

Looking at the correlation matrix and the hue I picked. I can clearly distinct two groups using positives and negatives. Even though the positives are greately dispersed, I will still group them into a unit. The two groups will be <0.09 (Negatives) and >=0.1 ( Posiitives)

```python
# This group has a correlation of <0.09
group_1 = whiskey[["Body","Floral", "Fruity", "Malty", "Honey", "Sweetness"]]

# This group has a correlation of >=0.1
group_2 = whiskey[["Body","Nutty", "Winey", "Tobacco", "Medicinal", "Spicy", "Smoky"]]
```

## For group 1.

```python
metrics = ['count', 'min', 'max', 'mean','std','skew']
```

```python
group_1.groupby(by='Body').agg(metrics)
```

```python
group_1.plot.hist(stacked=True,by='Body')
plt.show()
```

## For group 2

```python
group_2.groupby(by='Body').agg(metrics)
```

```python
group_2.plot.hist(stacked=True,by='Body')
plt.show()
```

## Clustering

```python
flav = []

for i in range(2,14):
    km = KMeans(n_clusters = i, n_init=5, max_iter = 500, random_state =0)
    km.fit(flavours)
    flav.append(km.inertia_)
    
plt.plot(range(2,14),flav)
plt.show()
```

```python
km = KMeans(n_clusters = i, n_init=5, max_iter = 500, random_state =0)
whiskey['type'] = km.fit_predict(flavours)
whiskey['type'].values
```
